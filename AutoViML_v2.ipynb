```python
import pandas as pd
from scipy import interpolate  # Changed from 'from scipy import interp'
from autoviml.Auto_ViML import Auto_ViML

# Define target variable
target = 'Estimated Revenue (USD)'

# Initialize AutoViML
automl = Auto_ViML()

# Fit the model
model, features, trainm, testm = automl.Auto_ViML(
    train=df,
    target=target,
    test='',
    sample_submission='',
    hyper_param='RS',
    feature_reduction=True,
    scoring_parameter='accuracy',
    KMeans_Featurizer=False,
    Add_Poly=0,
    Stacking_Flag=False,
    Binning_Flag=False,
    Imbalanced_Flag=False,
    verbose=1
)

# Print the selected features
print("Selected Features:", features)
```
